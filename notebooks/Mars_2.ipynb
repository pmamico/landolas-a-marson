A f√ºggv√©ny bemen≈ë param√©terei:<br>
	‚Ä¢	x: a haj√≥ aktu√°lis v√≠zszintes poz√≠ci√≥ja<br>

	‚Ä¢	fordulopont: az a poz√≠ci√≥, ahol ir√°nyt kell v√°ltani<br>
	‚Ä¢	leszallas_x: a lesz√°ll√°si z√≥na intervalluma, egy t√∂mb (pl. (90, 110))<br>

A teljes megold√°sban:<br>
	‚Ä¢	a haj√≥t forgasd balra vagy jobbra 30 fokkal, hogy k√∂zel√≠tsen a lesz√°ll√°si z√≥n√°hoz<br>

‚Ä¢ forgat√°s: a haj√≥ forgat√°s√°nak m√©rt√©ke (negat√≠v √©rt√©k balra, pozit√≠v √©rt√©k jobbra, 0 eset√©n nincs forgat√°s)<br>

## 7) Vez√©rl√©s

Ebben a r√©szben egy olyan f√ºggv√©nyt kell meg√≠rni, amely eld√∂nti,<br>
hogy a haj√≥nak **mit kell csin√°lnia az adott pillanatban**.<br>
Ez a f√ºggv√©ny lesz a ‚Äûf≈ë ir√°ny√≠t√≥‚Äù, ami meghat√°rozza a k√∂vetkez≈ë l√©p√©st.<br>
<br>
A feladat h√°rom r√©szre bonthat√≥:<br>
1) kezdeti stabiliz√°l√°s (ellens√∫lyoz√°s),<br>
2) mozg√°s a lesz√°ll√°si z√≥na f√∂l√©,<br>
3) biztons√°gos lesz√°ll√°s.<br>
<br>
---<br>
### 1) Kezdeti stabiliz√°l√°s (ellens√∫lyoz√°s)<br>
Ha a haj√≥ indul√°skor oldalir√°nyban t√∫l gyorsan mozog,<br>
el≈ësz√∂r ezt a sebess√©get kell lecs√∂kkenteni.<br>
<br>
Ha a bemeneti v√°ltoz√≥ azt jelzi, hogy **stabiliz√°l√°s sz√ºks√©ges**, akkor:<br>
‚Ä¢ √°ll√≠tsd a tol√≥er≈ët egy nagyobb √©rt√©kre (pl. 4), hogy legyen el√©g er≈ë a korrekci√≥hoz<br>
‚Ä¢ d√∂ntsd meg a haj√≥t arra az oldalra, amerre a v√≠zszintes sebess√©g mutat<br>
‚Ä¢ a d√∂nt√©s m√©rt√©ke legyen ar√°nyos azzal, mennyire kell lass√≠tani az oldalir√°ny√∫ mozg√°st<br>
<br>
Ha a haj√≥ k√∂zben **felfel√© mozog**, akkor:<br>
‚Ä¢ vedd vissza a tol√≥er≈ët egy kisebb √©rt√©kre (pl. 2), hogy ne emelkedjen tov√°bb<br>
<br>
---<br>
### 2) Mikor van k√©sz a stabiliz√°l√°s?<br>
Folyamatosan vizsg√°ld a v√≠zszintes sebess√©get.<br>
<br>
Ha az oldalir√°ny√∫ sebess√©g m√°r **nagyon kicsi** (pl. abszol√∫t√©rt√©kben 2 alatt van), akkor:<br>
‚Ä¢ hagyd abba a forgat√°st (forgatas = 0)<br>
‚Ä¢ jelezd, hogy a kezdeti stabiliz√°l√°s befejez≈ëd√∂tt<br>
<br>
Ezut√°n d√∂ntsd el, mi legyen a k√∂vetkez≈ë l√©p√©s:<br>
‚Ä¢ ha a haj√≥ m√°r a lesz√°ll√°si z√≥na f√∂l√∂tt van ‚Üí induljon a lesz√°ll√°si f√°zis<br>
‚Ä¢ ha m√©g nincs a z√≥na f√∂l√∂tt ‚Üí induljon a mozg√°si f√°zis<br>
<br>
Mozg√°si f√°zis eset√©n sz√°molj ki egy **fordul√≥pontot**,<br>
ami a jelenlegi poz√≠ci√≥ √©s a c√©lpont k√∂z√∂tt f√©l√∫ton helyezkedik el.<br>
<br>
---<br>
### 3) Mozg√°s √©s lesz√°ll√°s kezel√©se<br>
Ha a kezdeti stabiliz√°l√°s m√°r nem sz√ºks√©ges, akkor a f√ºggv√©ny feladata csak az,<br>
hogy tov√°bbadja az ir√°ny√≠t√°st a megfelel≈ë f√°zisnak:<br>
<br>
‚Ä¢ mozg√°si f√°zis eset√©n h√≠vd meg a mozg√°st vez√©rl≈ë f√ºggv√©nyt<br>
‚Ä¢ lesz√°ll√°si f√°zis eset√©n h√≠vd meg a lesz√°ll√°st vez√©rl≈ë f√ºggv√©nyt<br>
<br>
Ha egyik f√°zis sem akt√≠v, adj vissza egy biztons√°gos alap√°llapotot.<br>
<br>
---<br>
### Mit kell visszaadnia a f√ºggv√©nynek?<br>
A f√ºggv√©nynek mindig **hat √©rt√©ket** kell visszaadnia, ebben a sorrendben:<br>
<br>
1) forgatas ‚Äì mennyire kell elforgatni a haj√≥t (balra negat√≠v, jobbra pozit√≠v, 0 = nem forgatunk)<br>
2) toloero ‚Äì a be√°ll√≠tott tol√≥er≈ë √©rt√©ke (0‚Äì4)<br>
3) kezdeti_stabilizalas_szukseges ‚Äì jelzi, hogy kell-e m√©g stabiliz√°lni<br>
4) mozgas ‚Äì jelzi, hogy a haj√≥ mozg√°si f√°zisban van-e<br>
5) leszallas ‚Äì jelzi, hogy a haj√≥ lesz√°ll√°si f√°zisban van-e<br>
6) fordulopont ‚Äì az a v√≠zszintes poz√≠ci√≥, ahol a mozg√°si f√°zisban ir√°nyt kell v√°ltani<br>
<br>
Ez a f√ºggv√©ny teh√°t **nem egyetlen mozdulatot sz√°mol ki**,<br>
hanem mindig megmondja, hogy **mi legyen a k√∂vetkez≈ë teend≈ë**.<br>

````python
def vezerles(kezdeti_stabilizalas_szukseges, kezdeti_vizszintes_sebesseg, vizszintes_sebesseg, fuggoleges_sebesseg, szukseges_vizszintes_gyorsulas, leszallas_x, x, cel, mozgas, leszallas, fordulopont, pi):
    if kezdeti_stabilizalas_szukseges:
        toloero = 4
        irany = irany_sebesseg(kezdeti_vizszintes_sebesseg)
        try:
            alap_szog = int(math.asin(szukseges_vizszintes_gyorsulas / toloero) * 180 / pi)
        except Exception:
            alap_szog = 0
        forgatas = irany * (alap_szog + 2)
        if fuggoleges_sebesseg > 0:
            toloero = 2
        if abs(vizszintes_sebesseg) < 2:
            forgatas = 0
            kezdeti_stabilizalas_szukseges = False
            if leszallas_x[0] < x < leszallas_x[1]:
                leszallas = True
            else:
                mozgas = True
                fordulopont = (x + cel[0]) // 2
            if mozgas:
                f, t, mozgas, leszallas, fordulopont = mozgasi_fazis(x, vizszintes_sebesseg, fordulopont, leszallas_x)
                return f, t, kezdeti_stabilizalas_szukseges, mozgas, leszallas, fordulopont
            if leszallas:
                f, t = leszallasi_fazis(fuggoleges_sebesseg)
                return f, t, kezdeti_stabilizalas_szukseges, mozgas, leszallas, fordulopont
        return forgatas, toloero, kezdeti_stabilizalas_szukseges, mozgas, leszallas, fordulopont
    else:
        if mozgas:
            f, t, mozgas, leszallas, fordulopont = mozgasi_fazis(x, vizszintes_sebesseg, fordulopont, leszallas_x)
            return f, t, kezdeti_stabilizalas_szukseges, mozgas, leszallas, fordulopont
        if leszallas:
            f, t = leszallasi_fazis(fuggoleges_sebesseg)
            return f, t, kezdeti_stabilizalas_szukseges, mozgas, leszallas, fordulopont
        return 0, 4, kezdeti_stabilizalas_szukseges, mozgas, leszallas, fordulopont
````

````python
# asin(0) ‚Üí alap_szog = 0, irany = 1 (pozit√≠v kezdeti_vizsz), forgatas = 1*(0+2)=2, toloero=4
def irany_sebesseg(vizszintes_sebesseg):
    if vizszintes_sebesseg < 0:
        return -1
    else:
        return 1
assert vezerles(
    True,     # kezdeti_stabilizalas_szukseges
    1,        # kezdeti_vizszintes_sebesseg
    3,        # vizszintes_sebesseg  (>=2 ‚Üí nem stabiliz√°l le)
    -10,      # fuggoleges_sebesseg (<=0 ‚Üí toloero marad 4)
    0,        # szukseges_vizszintes_gyorsulas
    (20, 30), # leszallas_x
    0,        # x
    (100, 0), # cel
    False,    # mozgas
    False,    # leszallas
    10,       # fordulopont
    math.pi   # pi
) == (2, 4, True, False, False, 10)
print("‚úÖ A vez√©rl√©s alap tesztje szerint j√≥l m≈±k√∂dik ")
````

````python
# Felfel√© mozg√°s stabiliz√°l√°sa
assert vezerles(
    True,
    1,
    3,        # abs(vizszintes_sebesseg) >= 2 ‚Üí stabiliz√°l√°s marad
    1,        # fuggoleges_sebesseg > 0 ‚Üí toloero = 2
    0,
    (20, 30),
    0,
    (100, 0),
    False,
    False,
    10,
    math.pi
) == (2, 2, True, False, False, 10)
print("‚úÖ Vez√©rl√©s kezdeti felfel√© mozg√°s eset√©n ")
````

````python
#Kezdeti stabiliz√°l√°s LE√ÅLL: lass√∫ v√≠zszintes sebess√©g + x lesz√°ll√≥ z√≥n√°ban ‚Üí azonnal lesz√°ll√°si f√°zis

assert vezerles(
    True,
    1,
    0,        # abs(vizszintes_sebesseg) < 2 ‚Üí stabiliz√°l√°s le√°ll
    0,        # leszallasi_fazis(0) -> (0,2)
    0,
    (20, 30),
    25,       # x a leszallas_x tartom√°nyban
    (100, 0),
    False,
    False,
    10,
    math.pi
) == (0, 0, False, False, True, 10)
print("‚úÖ Azonnali lesz√°ll√°s")
````
