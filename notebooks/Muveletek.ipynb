# Részfeladatok a Mars landoláshoz

Ahhoz, hogy le tudjuk programozni a mars landolást, meghatároztunk részfeladatokat:

- két szám távolságának kiszámolása
- két koordináta távolságának kiszámítása
- vízszintes sebesség alapján a hajó irányának meghatározása
- a leszállási pont (cél) melyik oldalon van az űrhajóhoz képest?
- mozgás végrehajtása
- leszállás végrehajtása
